**Project**: Preliminary Analysis of Surf and Ultimate Prepaid Plans

***Date: August 8, 2025***

# **Work Plan**
The analysis will be conducted in the following stages:

- Data Loading and Initial Exploration: I will begin by loading the five provided datasets (users, calls, messages, internet, and plans) and examining their structure, data types, and initial content to identify any immediate issues.

- Data Preparation and Cleaning: This stage involves correcting data types, handling any errors or inconsistencies, and then aggregating the data. I will calculate, for each user on a monthly basis, the number of calls made, minutes used, messages sent, and the volume of data consumed. Finally, I will calculate the monthly revenue generated by each user.

- Exploratory Data Analysis (EDA): I will analyze the aggregated data to understand the typical behavior of customers on each plan. This includes calculating descriptive statistics (mean, variance, standard deviation) and creating visualizations like histograms to compare the usage patterns between 'Surf' and 'Ultimate' subscribers.

- Hypothesis Testing: I will conduct two key statistical tests to validate our findings:

- Test the hypothesis that the average revenue from the 'Ultimate' and 'Surf' plans is different.

- Test the hypothesis that the average revenue from users in the NY-NJ area differs from that of users in other regions.

- Conclusion: I will conclude with a summary of the key findings and provide a data-driven recommendation on which plan is more profitable, thereby informing future advertising strategies.
